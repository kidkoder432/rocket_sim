```python
import plotly.graph_objs as go

from importlib import reload

import landing_sim
reload(landing_sim)
from landing_sim import *

data = []
ms = []

for m in range(500, 700, 5):
    ms.append(m / 1000)
    data.append([])
    ts = []
    if m % 5 == 0: print(m, end=" ")
    for alt in range(300, 600, 5):
        if alt % 5 == 0: print(alt)
        mx = m / 1000
        ax = alt / 10
        ts.append(ax)
        data[-1].append(-sim(ax, mx))

    print()

```

```python
from ipywidgets import Output, VBox

print(len(data), len(data[0]), len(ms), len(ts))
fig = go.FigureWidget(go.Heatmap(x=ts, y=ms, z=data))
fig.update_layout(width=1000, height=600, autosize=False)

out = Output()


@out.capture(clear_output=True)
def onclick(trace, points, selector):
    m = points.ys[0] / 1000
    a = points.xs[0] / 10
    ts, ys, vs, ms, at = sim(a, m, True)
    
    print(f"Mass: {m} kg; Altitude: {a} ({at} sec delay) m; landing vel: {vs[-1]} m/s")
    
    plot = go.Figure(go.Scatter(x=ts, y=ys, mode='lines'))
    plot.add_scatter(x=ts, y=vs, mode="lines")
    plot.add_scatter(x=ts, y=ms, mode="lines")
    plot.add_vline(x=at)
    plot.add_vline(x=at + BURN_TIME)
    plot.show()

fig.data[0].on_click(onclick)

VBox([fig, out])
```
